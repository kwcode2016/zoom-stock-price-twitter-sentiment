```python
tweet_zm_2020_jan_to_apr = tweet_zm_df[(tweet_zm_df['date'] > '2020-1-1') & (tweet_zm_df['date'] <= '2020-5-1')]
# zm_prices_jan_to_apr = zoom_stock_prices_df[(zoom_stock_prices_df['date'] > '2020-1-1') & (zoom_stock_prices_df['date'] <= '2020-5-1')]

tweet_zm_2020_jan_to_apr.shape #26793,28

# range_numbers = [0,1,3,5,10,20,50]
range_numbers = [0,1,20]


colors = 'rgbcmyk'

def range_func(range_list):
    # return array of all the ranges 
    print ('range list:') 
    print (range_list)

    return_range_array = []
    for i in range(len(range_list)):
        try:
            print (f'from {range_list[i]} to {range_list[i+1]-1}')
            return_range_array.append((range_list[i], range_list[i+1]-1))
        except:
            print(f'last one: {range_list[i]} and above ')
            return_range_array.append((range_list[i], range_list[i] + 100000))
    
    return (return_range_array)


range_list = range_func(range_numbers)
print(f'return value from range_func: {range_list}')


# likeCount
category = 'likeCount'
color_i = 0
for tmp_i in range_list:
    tmp_df = tweet_zm_2020_jan_to_apr[(tweet_zm_2020_jan_to_apr[category] > tmp_i[0]) & (tweet_zm_2020_jan_to_apr[category] <= tmp_i[1]  )].groupby([tweet_zm_2020_jan_to_apr.date.dt.date]).sum()
    # tmp_df.likeCount.plot(x='date', y="tweet number", figsize=(50,10), label=str(tmp_i))
    # tmp_df[category].plot(x='date', y=['a','b','c','d'], figsize=(50,20), label=str(tmp_i), color=colors[color_i], kind='bar', stacked=False)
    tmp_df[category].plot(x='date', y=['a','b','c','d'], figsize=(50,20), label=str(tmp_i), color=colors[color_i], marker='o')
    color_i += 1





import matplotlib.dates as mdates
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
plt.gcf().autofmt_xdate()
plt.title(category, fontsize=40)
plt.legend()
```
